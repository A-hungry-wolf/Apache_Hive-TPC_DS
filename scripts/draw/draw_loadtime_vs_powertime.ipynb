```python
import pandas as pd
# Load the Excel file
excel_file_path = '/content/LoadTimevsPowerTime.xlsx'
excel_df = pd.read_excel(excel_file_path)

# Display the first few rows of the dataframe
excel_df.head()
```

```python
import matplotlib.pyplot as plt

# Plotting
plt.figure(figsize=(10, 6))

# Plotting Load Test and Power Test against Scale Factors
plt.plot(excel_df['Scale Factors'], excel_df['Load Test'], label='Load Test', marker='o', color='blue')
plt.plot(excel_df['Scale Factors'], excel_df['Power Test'], label='Power Test', marker='x', color='red')

# Adding labels, title and legend
plt.xlabel('Scale Factors')
plt.ylabel('Time (seconds)')
plt.title('Comparison of Load Test and Power Test Times across Scale Factors')
plt.legend()

# Making the plot look academic and professional
plt.grid(True)
plt.xticks(excel_df['Scale Factors'])
plt.tight_layout()

# Show the plot
plt.show()
```
